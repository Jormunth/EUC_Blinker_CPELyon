```python
filename = "turn2.csv"
import pandas as pd
import numpy as np

columns = ["ts","y","p","r"]
df = pd.read_csv(filename, skiprows=1, names=columns)

df = df[["y","p","r"]][:]

# Unwrap angular data
df["y"] = np.unwrap(df["y"])
df["p"] = np.unwrap(df["p"])
df["r"] = np.unwrap(df["r"])

df.plot()
```
